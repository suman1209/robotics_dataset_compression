```python
def get_vid2img(video_path, output_dir, fps=30, img_format="frame_%03d.png"):

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    command = f"ffmpeg -i {video_path} -vf fps={fps} {os.path.join(output_dir, img_format)}"
    os.system(command)
    
    # Count the number of extracted frames
    num_frames = len([f for f in os.listdir(output_dir) if f.endswith(".png")])
    print(f"Extracted {num_frames} frames from {video_path} to {output_dir}")
    return num_frames


    
```

```python
import os
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim
from natsort import natsorted
import cv2
import numpy as np

def compare_images(img1_path, img2_path):
    """
    Compare two images using MSE and SSIM, handling common issues.
    
    Args:
        img1_path (str): Path to the first image.
        img2_path (str): Path to the second image.
    
    Returns:
        tuple: MSE and SSIM values.
    """
    img1 = cv2.imread(img1_path, cv2.IMREAD_COLOR)
    img2 = cv2.imread(img2_path, cv2.IMREAD_COLOR)

    # Normalize ANd MSE
    img1 = img1.astype("float32") / 255.0
    img2 = img2.astype("float32") / 255.0
    mse = np.mean((img1 - img2) ** 2)

    # Compute SSIM (Similarity Index [0,1]
    win_size = min(7, img1.shape[0], img1.shape[1])
    ssim_value = ssim(img1, img2, channel_axis=-1, win_size=win_size, data_range=1.0)

    return mse, ssim_value

#compare all images in 2 folders
def compare_datasets(original_dir, new_dir, output_report="comparison_report.txt"):
    
    original_images = natsorted(
        [f for f in os.listdir(original_dir) if f.endswith(".png") and "resize" not in f]
    )
    new_images = natsorted(
        [f for f in os.listdir(new_dir) if f.endswith(".png")]
    )

    if len(original_images) != len(new_images):
        print(f"Warning: Number of images differ ({len(original_images)} vs {len(new_images)})")

    report = []
    for orig_img, new_img in zip(original_images, new_images):
        orig_path = os.path.join(original_dir, orig_img)
        new_path = os.path.join(new_dir, new_img)
        
        try:
            mse, ssim_value = compare_images(orig_path, new_path)
            report.append(f"{orig_img} vs {new_img}: MSE={mse:.2f}, SSIM={ssim_value:.4f}")
        except ValueError as e:
            report.append(f"{orig_img} vs {new_img}: ERROR - {str(e)}")
    
    # Save the report
    with open(output_report, "w") as f:
        f.write("\n".join(report))
    
    print(f"Comparison report saved to {output_report}")
```

MSE = Pixel by pixel difference
SSIM: Structural Similarity Index (SSIM) measures the similarity between two images by comparing luminance, contrast, and structural differences, providing a score between 0 (no similarity) and 1 (identical).
