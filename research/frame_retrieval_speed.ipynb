```python
# import shutil

import os
from math import ceil

import os
import shutil
from math import ceil

def chunk_images_to_directories(image_dir, output_dir, chunk_size):
    """
    Organize images from a directory into multiple subdirectories in order.
    
    Args:
        image_dir (str): Path to the directory containing images.
        output_dir (str): Path to the base output directory where subdirectories will be created.
        chunk_size (int): Number of images per chunk (subdirectory).
    """
    # Get list of all image files
    all_images = [f for f in os.listdir(image_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff'))]
    all_images.sort()  # Optional: Keep consistent order

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Divide images into chunks and move to subdirectories
    num_chunks = ceil(len(all_images) / chunk_size)
    for i in range(num_chunks):
        chunk = all_images[i * chunk_size:(i + 1) * chunk_size]
        
        # Create a subdirectory for the chunk
        chunk_dir = os.path.join(output_dir, f'chunk_{i + 1}')
        os.makedirs(chunk_dir, exist_ok=True)

        # Move files to the chunk directory
        for i, img_file in enumerate(chunk):
            src_path = os.path.join(image_dir, img_file)
            dst_path = os.path.join(chunk_dir, f"frame_{i:03}.png")
            shutil.copy(src_path, dst_path)  # Use shutil.copy() if you want to keep the original files

    print(f"Images organized into {num_chunks} directories in '{output_dir}'.")

# Example usage
image_directory = '/var/lit2425/jenga/suman/robotics_dataset_compression/datasets/extracted_frames_video_0'
output_directory = '/var/lit2425/jenga/suman/robotics_dataset_compression/datasets/'
num_frames = 851
num_parts = 40
num_frames_per_part = num_frames // num_parts
chunk_size = num_frames_per_part

chunk_images_to_directories(image_directory, output_directory, chunk_size)


# # Example usage
# image_directory = '/path/to/your/image/directory'
# num_frames = 851
# num_parts = 2
# num_frames_per_part = num_frames // num_parts
# chunk_size = num_frames_per_part
# chunks = chunk_directory(image_directory, chunk_size)


# chunk_images_to_directories(image_directory, output_directory, chunk_size)

# part = 0
# 
# for i in range(num_frames):
#     if i > num_frames_per_part:
#         part += 1
#     shutil.copy(f"/var/lit2425/jenga/suman/robotics_dataset_compression/datasets/extracted_frames_video_0{i}.png", f"./part{part}/idx_{i-num_frames_per_part}.png")
```

```python
# the following code is obtained from piccomparision notebook in our repo/research
import os

def extract_frames_from_video(video_path, output_dir, fps=30, img_format="frame_%03d.png"):

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    command = f"ffmpeg -i {video_path} -vf fps={fps} {os.path.join(output_dir, img_format)}"
    os.system(command)
    
    # Count the number of extracted frames
    num_frames = len([f for f in os.listdir(output_dir) if f.endswith(".png")])
    print(f"Extracted {num_frames} frames from {video_path} to {output_dir}")
    return num_frames
    

i = 0
output_dir = f"./extracted_frames_video_{i}"
num_frames = extract_frames_from_video(video_path, output_dir, fps=10)
print(f"Video {i}: Extracted {num_frames} frames.")
```
